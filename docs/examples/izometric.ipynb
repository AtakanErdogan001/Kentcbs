```python
import matplotlib.pyplot as plt
import prettymaps

# Create the plot
plot = prettymaps.plot(
    'Stad van de Zon, Heerhugowaard, Netherlands',
    preset='minimal'
)

# Set equal aspect ratio
plt.gca().set_aspect('equal')

# Rotate the plot
plt.gca().set_xticks([])
plt.gca().set_yticks([])
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import prettymaps
from PIL import Image

# Create the plot
plot = prettymaps.plot(
    'Stad van de Zon, Heerhugowaard, Netherlands',
    preset='minimal'
)

# Save the plot as an image file
plot.savefig('prettymaps_plot.png')

# Close the plot to release resources
plt.close(plot)

# Load the saved image with PIL
plot_pil = Image.open('prettymaps_plot.png')

# Resize the image to achieve an isometric-like effect
plot_pil_resized = plot_pil.resize((plot_pil.width // 2, plot_pil.height // 2))  # Adjust the scaling factor as needed

# Show the resized plot
plot_pil_resized.show()
```
